```python
import numpy as np
import pandas as pd
```

```python
#criar dataframe aleatorio contendo nan
def data(size,nan):
    df = np.random.rand(size)
    df.ravel()[np.random.choice(df.size,nan,replace=False)] = np.nan
    return df
dfs = [None]*5
for i in range(5):
    dfs[i] = pd.DataFrame(data(100,int(np.random.randint(30))),columns=['df'+str(i)+'_'])
dfs
```

```python
#preencher valores nan com a m√©dia
def fill(data,value):
    data = data.fillna(value)
    return data
for i in range(len(dfs)):
    if dfs[i].isnull().values.any() == True:
        #------------------------------------------------
        #usar este se for substituir a dataframe antiga
        #dfs[i] = fill(dfs[i],dfs[i].mean())
        #------------------------------------------------
        dfs[i]['dfs'+str(i)+'_after'] = fill(dfs[i],dfs[i].mean())
        print(dfs[i])
```

```python
#mapeamento utilizando dicionario
gender = {0:'male',1:'female'}
#random dataframe with 0 or 1
data = pd.DataFrame(np.random.choice([0, 1], size=100),columns=['gender'])
print(data.head(5),'\n')
data['gender'] = data['gender'].map(gender)
print(data.head(5))
```

```python
#mapeamento utilizando def
def color_func(val):
    if val <= 200:
        return 'red'
    elif val <=400:
        return 'yellow'
    elif val <=600:
        return 'green'
    elif val <=800:
        return 'blue'
    else:
        return 'purple'
data = pd.DataFrame(np.random.choice(np.arange(1000), size=100),columns=['color'])
print(data.head(10),'\n')
data['color'] = data['color'].map(color_func)
print(data.head(10))
```

```python
#separar par e impar
def par(data):
    return list(filter(lambda x: x%2==0, data))
def impar(data):
    return list(filter(lambda x: x%2!=0, data))
data = np.random.choice(np.arange(1000), size=100)
data = pd.DataFrame([par(data),impar(data)]).T
data.columns = ['par', 'impar']
data
```
